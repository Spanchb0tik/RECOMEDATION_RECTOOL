```python
from rectools import Columns
from rectools.dataset import Dataset
import pandas as pd

data = pd.read_csv(
    "/content/df_temp.csv",
    engine="python",
    header = None,
    names=[Columns.Item, Columns.User, Columns.Weight, Columns.Datetime],
)
```

```python
data['datetime'] = pd.to_datetime(data['datetime'])
```

```python
dataset = Dataset.construct(data)
```

```python
from rectools.models import PureSVDModel

# Fit model and generate recommendations for all users

model = PureSVDModel()
model.fit(dataset)
recos = model.recommend(
    users=data[Columns.User].unique(),
    dataset=dataset,
    k=7,
    filter_viewed=True,
)
```

```python
data.user_id.value_counts()
```

```python
alls = recos[['user_id', 'item_id']].groupby(by=["user_id", 'item_id']).first(7)
```

```python
users = {}

for user_id, hashi in alls.index:
  users[user_id] = users.get(user_id, []) + [hashi]
```

```python
tmp = pd.DataFrame(users.keys())
tmp2 = pd.DataFrame(users.values())

df_000 = pd.concat([tmp, tmp2], axis = 1)
df_000.columns = ['customer_id'] + [f"community_id_{i}" for i in range(1, 8)]
```

```python
res_table = df_000[df_000['customer_id'].isin(sample.customer_id.values)]
```
